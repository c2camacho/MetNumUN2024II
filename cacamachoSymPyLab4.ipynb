```
from sympy import symbols, I, N, pi, expand
x1, x2, x3, y1, y2, y3, z1, z2, z3 = symbols("x1 x2 x3 y1 y2 y3 z1 z2 z3", real=True)
x = x1+I*x2+I*x3
y = I*y1-y2+y3
z = z1*z2+I*z3

print(x+y*z)
print(expand(x+y*z))
w=3/(pi + I)**2
print('w=',w)
w=N(w)
print('w=',w)
```
